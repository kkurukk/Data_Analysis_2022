```python
# Another method to resolve the question above is to create a function
def getDate(data):
    data = str(data)
    year = data[0:4]
    month = data[4:6]
    day = data[6:8]
    return year + '-' + month + '-' + day

data['사용일자1'] = data['사용일자'].apply(getDate)
data.head()
```
